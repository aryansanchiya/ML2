```python
import pandas as pd
```

```python
bollywood_movies = pd.read_csv("bollywood_meta_2010-2019.csv") 
```

```python
movie_ratings = pd.read_csv("bollywood_ratings_2010-2019.csv")
```

```python
frames = [bollywood_movies,movie_ratings]
```

```python
main_data = pd.concat(frames)
```

```python
print(main_data)
```

```python
main_data.info()
```

```python
#How many Movies got realeased in each genre?
main_data.genres.value_counts()
```

```python
#How many movies in each genre got released in different release times like long weekend, festive seasons,etc.
pd.crosstab(main_data['year_of_release'],main_data['genres'])
```

```python
#Which year maximum movies got realease?
main_data.year_of_release.value_counts()
```
