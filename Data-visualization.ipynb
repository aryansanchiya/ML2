```python
#Distribution Plot
sn.displot(ipl_auction["Auction_Price"])
```

## Box Plot
box = sn.boxplot(ipl_auction['T20 caps'])

```python
#Comparing Distributions
sn.distplot(ipl_auction[ipl_auction['T20 caps']>=1]['Auction_Price'],
            color = 'g',
            label = 'Captaincy Experience'
           );
sn.distplot(ipl_auction[ipl_auction['T20 caps']<1]['Auction_Price'],
            color = 'r',
           label = 'No Captaincy Experience'
          );
plt.legend();
```

```python
#Comparing Distributions
sn.histplot(ipl_auction[ipl_auction['T20 caps']>=1]['Auction_Price'],
            color = 'g',
            label = 'Captaincy Experience'
           );
sn.histplot(ipl_auction[ipl_auction['T20 caps']<1]['Auction_Price'],
            color = 'r',
           label = 'No Captaincy Experience'
          );
plt.legend();
```
