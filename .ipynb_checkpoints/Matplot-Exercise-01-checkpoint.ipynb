```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
%matplotlib inline
```

```python
main_data = pd.read_csv("World_Population.csv")
```

```python
main_data
```

```python
#Draw to histogram of migrants
plt.hist(main_data['Migrants (net)'],bins=30)
```

```python
#Draw the distribution plot for 
sn.displot(main_data['Net Change'])
```

```python
main_data
```

```python
#Compare the distributions of the "fertility rate of country".
sn.distplot(main_data[main_data['Density(P/Km²)']>250]['Land Area(Km²)'],
           color = "r",
           label = "Migrants in Positive");

sn.distplot(main_data[main_data["Density(P/Km²)"]<250]['Land Area(Km²)'],
           color = "g",
           label="Migrants in Negative");
plt.legend();

```
