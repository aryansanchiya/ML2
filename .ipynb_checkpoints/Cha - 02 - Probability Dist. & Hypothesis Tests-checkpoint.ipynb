# Probability  Mass Function:

<h6>Here's the PMF formula:
    <br>

<i>P(X=xi)=pi</i></h6>


<h6>In simpler terms, the PMF tells us the likelihood of each possible outcome of a random variable.

Let's take an example to illustrate this. Suppose we have a fair six-sided die. The random variable XX represents the outcome of rolling the die.

    The possible outcomes are x1=1, x2=2, x3=3, x4=4, x5=5, and x6=6.
    Since the die is fair, each outcome has an equal probability of 1/6.

So, the PMF for this scenario is:

P(X=1)=1/6
P(X=2)=1/6
P(X=3)=1/6
P(X=4)=1/6
P(X=5)=1/6
P(X=6)=1/6
</h6>

# Binomial Distribution



n simpler terms, the formula calculates the probability of getting exactly kk successes in nn independent trials, where each trial has a probability pp of success.




```python
#Example 3.4.1 (P.No : 66)
#1.Probability that exactly 5 customers will return the items.

from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
```

```python
stats.binom.pmf(5,20,0.1)
```

```python
pmf_df = pd.DataFrame({'success': range(0,21),
                      'pmf': list(stats.binom.pmf(range(0,21),
                                                 20,0.1))})

sn.barplot(x = pmf_df.success, y = pmf_df.pmf)
plt.ylabel('pmf')
plt.xlabel("Number of items returned");
```

```python
stats.binom.cdf(5,20,0.1)
```

<h1>Poisson Distribution </h1>

<h6>What is Poisson Distribution?</h6>
<h6>Ans:- The Poisson distribution is a probability distribution that helps us model the number of times an event occurs within a fixed interval of time or space<h6>
    
<h6>-->It's particularly useful for events that happen independently and with a known constant average rate.<h6>

# 

<h5>Parameters

    λ (lambda): The average number of events in the given interval. It's the key parameter of the Poisson distribution.</h5>

# 

# 

# 

```python
#Number of calls arriving at call center : 10 calls per hour
#1)Calculate the probability that number of calls will be maximum 5:
stats.poisson.cdf(5,10)
```

```python
#2)Calculate the probability that the number of calls over a 3 hour period will be exceeded 30:1-
1 - stats.poisson.cdf(30,30)
```

```python
#Range(0,30) returns all values from 0 to 30 (excluding 30)
pmf_df = pd.DataFrame({'success':range(0,30), 'pmf':list(stats.poisson.pmf(range(0,30),10))})
sn.barplot(x=pmf_df.success,y=pmf_df.pmf)
plt.xlabel("Number of calls Received");
```

# <h1>Exponential Distribution

# <h5>Exponential Distribution is a single parameter continuous distrbution that is traditionally used for modeling time-to-failure of electronic components. The exponential distribution represents a process in which event occur continuously and independently at a constant average rate.

# 

where lambda = rate of occurence of the event.
Mean of exponential distribution is given by 1/lambda.

```python
#1) The probability that the system will fail before 1000 hours:

stats.expon.cdf(1000,
               loc=1/1000,
               scale=1000)
```

```python
#2) Probability that it will not fail upto 2000 hours:
1 - stats.expon.cdf(2000,
               loc=1/1000,
               scale=1000)
```

```python
#3) Calculate the time by which 10% system will falie. ppf(Percent Point Function)
stats.expon.ppf(.1,
               loc=1/1000,
               scale=1000)
```

# Cumulative Probability Distribution:

```python
#PROBABILITY THAT LOSS WILL BE MORE THAN 2%
print("Proba. of making 2% loss or more: ")
stats.norm.cdf(-0.02,
              loc=glaxo_df.gain.mean(),
              scale=glaxo_df.gain.std())
```

```python
print("Proba. of making 2% loss or more: ")
stats.norm.cdf(-0.02,
              loc=beml_df.gain.mean(),
              scale=beml_df.gain.std())
```

```python
#WHICH STOCK HAS HIGHER PRIORITY OF MAKING A DAILY RETURN OF 2% OR MORE
print("Proba. of making 2% PROFIT or more OF GLAXO: ",
1 - stats.norm.cdf(0.02,
              loc=glaxo_df.gain.mean(),
              scale=glaxo_df.gain.std()))
```

```python
print("Proba. of making 2% PROFIT or more OF BEML: ",
1 - stats.norm.cdf(0.02,
              loc=beml_df.gain.mean(),
              scale=beml_df.gain.std()))
```
